# Bivariate and multivariate analysis 
In this workbork we will see how we can do EDA through Bivariate and Multivariate

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

### Import data set

```python
tipps_df = sns.load_dataset('tips')
tipps_df.head()
```

## Scatterplot (Numerical - Numerical)

```python
## Scatterplot (Numerical - Numerical)
sns.scatterplot(data=tipps_df, x='total_bill', y='tip' , hue='sex')
plt.show()
```

## Barplot ( Numerical - Categorical)

```python
titanic_df = sns.load_dataset('titanic')
```

```python
# barplor ( Numerical - Categorical)  See how age varies with class
sns.barplot(data=titanic_df, x='class', y='age')
```

## BoxPlot ( Numerical - categorical)

```python
sns.boxplot(data=titanic_df, x='sex', y ='age', hue='survived')
plt.show()
```

## DistPlot ( Numerical - Categorical)

```python
sns.displot(data=titanic_df, x='sex', hue='survived')
plt.show()
```

## Heatmap ( Categorical - categorical)

```python
# Create crosstab
ct = pd.crosstab(index=titanic_df['pclass'], columns=titanic_df['survived'])
ct

```

```python
# Create heatmap from crosstab
plt.figure(figsize=(8, 6))
sns.heatmap(ct, annot=True, fmt='d', cmap='YlOrRd')
```

## Clustermap ( Categorical - Categorical)

```python
pd.crosstab(index=titanic_df['sibsp'], columns=titanic_df['survived'])
```

```python
sns.clustermap(pd.crosstab(index=titanic_df['parch'], columns=titanic_df['survived']))
plt.show()
```

## pairPlot

```python
## pair plot
iris_df = sns.load_dataset('iris')
sns.pairplot(iris_df)
```

## LinePlot
Use when you have date or year or like this x-exes


```python
flight_df = sns.load_dataset('flights')
flight_df.head()
```

```python
new = flight_df.groupby('year')['passengers'].sum().reset_index()
new
```

```python
sns.lineplot(data=new, x='year', y='passengers')
```

```python
sns.heatmap(flight_df.pivot(index='month', columns='year', values='passengers'))
```

```python
sns.clustermap(flight_df.pivot(index='month', columns='year', values='passengers'))
```
